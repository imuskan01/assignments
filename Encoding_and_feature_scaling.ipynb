```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, RobustScaler

# Step 1: Create Dataset
data = {
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female'],
    'City': ['Mumbai', 'Delhi', 'Chennai', 'Mumbai', 'Delhi'],
    'Size': ['Small', 'Large', 'Medium', 'Large', 'Small'],
    'Age': [25, 45, 32, 60, 28],
    'Fare': [15, 300, 85, 450, 20]
}

df = pd.DataFrame(data)

print("Original Dataset:\n", df)


# Task 1 — Categorical Encoding

# 1. Gender → Label Encoding
le_gender = LabelEncoder()
df['Gender'] = le_gender.fit_transform(df['Gender'])

# 2. City → One-Hot Encoding (drop one column to avoid dummy variable trap)
df = pd.get_dummies(df, columns=['City'], drop_first=True)

# Small = 0, Medium = 1, Large = 2
size_mapping = {'Small': 0, 'Medium': 1, 'Large': 2}
df['Size'] = df['Size'].map(size_mapping)

print("\nAfter Encoding:\n", df)

# Task 2 — Feature Scaling

scaler = RobustScaler()

# Apply scaling to Age and Fare
df[['Age', 'Fare']] = scaler.fit_transform(df[['Age', 'Fare']])

print("\nAfter Scaling:\n", df)
```
